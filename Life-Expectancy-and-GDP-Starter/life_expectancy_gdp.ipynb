# Life Expectancy and GDP.
For this project, we will analyze data on GDP and life expectancy from the World Health Organization and the World Bank to try and identify the relationship between the GDP and life expectancy of six countries(Chile, China, Germany,Mexico, United States and Zimbawe). The available data includes the years 2000 trough 2015. The available columns from the subset of data includes. <br>

1.  `Country`
2.  `Year`
3.  `Life expectancy at birth (years)` `LEABY`
4.  `GDP`

_GDP Source:World Bank national accounts data, and OECD National Accounts data files._

### These are the question we will answer.

- Has life expectancy increased over time in the six nations?
- Has GDP increased over time in the six nations?
- Is there a correlation between GDP and life expectancy of a country?
- What is the average life expactancy in these nations?
- What is the distribution of that life expectancy?




Import libaries and load the available data from `'all_data.csv'` into a dataframe `df`.
Rename column `'Life expectancy at birth (years)'` to `'LEABY'`

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load all_data.csv into the DataFrame df
df = pd.read_csv('all_data.csv')
# Check the data and column names
print(df)

# Rename colum Life expectancy at birth (years) to LEABY to be more manageable
df.rename(columns={'Life expectancy at birth (years)' : 'LEABY'}, inplace = True)
# Check if the column was renamed
print(list(df.columns))
```

## Life expectancy over time for the six nations.
Create function to plot the information



```python
# Create plot function for country's LEABY or GDP by Year
def plot_country_data(df, ax, country, xcol, ycol):
    
    plt.plot(df[xcol], df[ycol])
    ax.set_title(country)
    ax.set_xlabel(xcol)
    ax.set_ylabel(ycol)
    if (ycol == 'LEABY'):
        # Set Y axis range
        plt.ylim([42,82])
    else:
        #plt.ylim([1000000000, 19000000000000])
        plt.ylabel('GDP in USD')   
    
```

 **Create dataframes for each of the six countries**

```python
# Get the unique list of countries
countries_list = set(df.Country)
# Sort countries alphabetically
countries_list = sorted(countries_list)
print(countries_list)

chile_df = df[df.Country == 'Chile']
china_df = df[df.Country == 'China']
germany_df = df[df.Country == 'Germany']
mexico_df = df[df.Country == 'Mexico']
usa_df = df[df.Country == 'United States of America']
zimbabwe_df = df[df.Country == 'Zimbabwe']

```

Plot graphics for each of the six countries in a figure with two rows and three columns.

```python
# I do this to get big fonts in the plot
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("bright")

plt.figure(figsize=(25,15))
# Create 6 subplots 2 rows and 3 columns
ax = plt.subplot(2, 3, 1)
# Call plot function by country 
plot_country_data(chile_df, ax, 'Chile', 'Year', 'LEABY')

ax = plt.subplot(2, 3, 2)
# Call plot function by country 
plot_country_data(china_df, ax, 'China', 'Year', 'LEABY')

ax = plt.subplot(2,3,3)
# Call plot function by country 
plot_country_data(germany_df, ax, 'Germany', 'Year', 'LEABY')

ax = plt.subplot(2, 3, 4)
# Call plot function by country 
plot_country_data(mexico_df, ax, 'Mexico', 'Year', 'LEABY')

ax = plt.subplot(2, 3, 5)
# Call plot function by country 
plot_country_data(usa_df, ax, 'United States of America', 'Year', 'LEABY')

ax = plt.subplot(2, 3, 6)
# Call plot function by country 
plot_country_data(zimbabwe_df, ax, 'Zimbabwe', 'Year', 'LEABY')
# Create some space between plots
plt.subplots_adjust(wspace=.3, hspace=.3)
# Save plot
plt.savefig('LEABY by Country.png')
plt.show()
```

### Compare all countries life expectancy in one graph for better visualization.


```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("colorblind")
plt.figure(figsize=(18,12))
sns.lineplot(x=df.Year, y=df.LEABY, hue=df.Country)
#loc='center left', bbox_to_anchor=(1, 0.5), ncol=1
plt.legend()
plt.ylabel('Life Expectancy at Birth in Years')
plt.title('Comparative Life Expectancy per Country')
plt.savefig('Summarized LEABY by Country.png')
plt.show()
```

The charts above shows the life expectancy `LEABY` over the years. The chart shows that every country has been increasing their life expectancy, but Zimbabwe has seen the greatest increase after a dip around 2004.

### Has GDP increased over time in the six nations?
Let plot the graphics of the `GDP` over time for each country

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("RdBu")

plt.figure(figsize=(25,15))
ax = plt.subplot(2, 3, 1)
# Call plot function by country 
plot_country_data(chile_df, ax, 'Chile', 'Year', 'GDP')

ax = plt.subplot(2, 3, 2)
# Call plot function by country 
plot_country_data(china_df, ax, 'China', 'Year', 'GDP')

ax = plt.subplot(2,3,3)
# Call plot function by country 
plot_country_data(germany_df, ax, 'Germany', 'Year', 'GDP')

ax = plt.subplot(2, 3, 4)
# Call plot function by country 
plot_country_data(mexico_df, ax, 'Mexico', 'Year', 'GDP')

ax = plt.subplot(2, 3, 5)
# Call plot function by country 
plot_country_data(usa_df, ax, 'United States of America', 'Year', 'GDP')

ax = plt.subplot(2, 3, 6)
# Call plot function by country 
plot_country_data(zimbabwe_df, ax, 'Zimbabwe', 'Year', 'GDP')
plt.subplots_adjust(wspace=.3, hspace=.3)
plt.savefig('GDP by Country.png')
plt.show()
```

This chart will compare the `GDP` growth for each country in on graphic.

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("colorblind")
plt.figure(figsize=(18,12))
sns.lineplot(x=df.Year, y=df.GDP, hue=df.Country)
plt.legend()
plt.ylabel("GDP in Trillions of U.S. Dollars")
plt.title('Comparative GDP Growth per Country')
plt.savefig('Summarized GDP by Country.png')
plt.show()
```

In these plots we can see that the US and China have had a significant increase in `GDP` throughout the year. Germany has a modest growth. Mexico and Chile had a very minimal growth and Zimbabwe has been steady no major change.


### Is there a correlation between GDP and life expectancy of a country?

Create a function to scatter the data by `Country`.

```python
def scatter_country_gdp_LEABY(df, ax, country):
    labels = list(df.Year.values)
    str_labels = list(map(str,labels))
    plt.scatter(df.GDP, df.LEABY, marker = 'o', c= df.Year, label= str_labels)
    #plt.legend(str_labels)
    ax.set_title(country)
    ax.set_xlabel('GDP in USD')
    ax.set_ylabel('LEABY')
    plt.subplots_adjust(wspace=.3, hspace=.3)
   
```

Call function for each of the countries and set different subplots.

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("deep")

fig = plt.figure(figsize=(25,15))
ax = plt.subplot(2, 3, 1)
# Call scatter function by country 
scatter_country_gdp_LEABY(chile_df, ax, 'Chile')

ax = plt.subplot(2, 3, 2)
# Call scatter function by country
scatter_country_gdp_LEABY(china_df, ax, 'China')

ax = plt.subplot(2,3,3)
# Call scatter function by country 
scatter_country_gdp_LEABY(germany_df, ax, 'Germany')

ax = plt.subplot(2, 3, 4)
# Call scatter function by country 
scatter_country_gdp_LEABY(mexico_df, ax, 'Mexico')

ax = plt.subplot(2, 3, 5)
# Call scatter function by country 
scatter_country_gdp_LEABY(usa_df, ax, 'United States of America')

ax = plt.subplot(2,3,6)
scatter_country_gdp_LEABY(zimbabwe_df, ax, 'Zimbabwe')
plt.savefig('Distribution LEABY vs GDP.png')
plt.show()
```

The following is a scatter plot with all the countries comparing their `GDP` and `LEABY` for better visualization.

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("deep")
plt.figure(figsize=(25,15))
sns.scatterplot(x=df.LEABY, y=df.GDP, hue=df.Country).legend()
plt.title('GDP and Life Expectandy Distribution by Country')
plt.savefig('LEABY vs GDP Distribution Summarized.png')
plt.show()
```

The previous two charts showed the relationship between `GDP` and `LEABY` (Life Expectancy at Birth in Years). It looks like the previous charts where `GDP` for Zimbabwe is staying flat, while their life expectancy is going up. For the other countries they seem to exhibit a rise in life expectancy as GDP goes up. The US and China seem to have very similar slopes in their relationship between `GDP` and `LEABY`.

### What is the average life expectancy in these nations?

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("pastel")

plt.figure(figsize=(20,10))
sns.barplot(x = "Country", y = "LEABY", data = df)
plt.title('Average Life Expectancy by Country')
plt.savefig('Average LEABY by Country.png')
plt.show()
```

The previous barplot shows the average `LEABY` by `Country`. Below we calculated and printed the averages by `Country`.

```python
avg_LEABY_by_country = df.groupby('Country').LEABY.mean()
print(avg_LEABY_by_country)
```

## What is the distribution of that life expectancy?

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("bright")

plt.figure(figsize=(20,10))
sns.violinplot(data=df, x="LEABY", y="Country")
plt.title('Life Expectancy Distribution by Country')
plt.xlabel('Life Expectancy at Birth in Years')
plt.savefig('LEABY Distribution by Country.png')
plt.show()
```

In the previous plot `LEABY` (Life Expectancy Distribution by Country), all of the countries have shorter ranges except for Zimbabwe which has a long range spanning from the high 30s to the high 60s.

## What is the average GDP of each country?

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("dark")

plt.figure(figsize=(20,10))
sns.barplot(x = "Country", y = "GDP", data = df)
plt.title('Average GDP by Country')
plt.ylabel('GDP in Trillions of U.S. Dollars')
plt.savefig('Average GDP by Country.png')
plt.show()
```

In this bar plot we can appreciate the average `GDP` differences between the six countries. With US having the highest, China second, Germany third, Mexico fourth, Chile fifth and Zimbabwe at the end.

## What is the GDP distribution for each country?

```python
sns.set_context('poster')
sns.set_style("darkgrid")
sns.set_palette("pastel")

plt.figure(figsize=(20,10))
sns.violinplot(data=df, x="GDP", y="Country")
plt.title('GDP Distribution by Country')
plt.xlabel('GDP in Trillions of USD')
plt.savefig('GDP Distribution by Country.png')
plt.show()
```

In the `GDP` distribution plot above, China and the US have a relatively wide ranges, where as Chile and Mexico have shorter ranges and Zimbabwe has almost no range.
