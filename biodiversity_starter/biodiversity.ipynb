### ELIMINATE DUPLICATES

Create a dataframe where we will store duplicates in the `scientific_name` column.

Then identify any possible conflicts with different `conservation_status` values.

```python

s_duplicates = spe_df[spe_df.scientific_name.duplicated(keep=False)]
print(f'Duplicates: {len(s_duplicates)}')
# Create a dataframe with the count of duplicate rows by scientific_name
duplicate_count = s_duplicates.groupby(['scientific_name']).conservation_status.count()
# Print the scientific_names with the highest duplicate counts and review the conservation_status
for name, status in duplicate_count.iteritems():
    if int(status) > 2:
        display(s_duplicates[s_duplicates.scientific_name == name])
```

In this report we see multiples duplicates with at least 3 values. The only one of concern is *Canis lupus* since it has 2 possible values. We will make sure we pick the most dangerus status of the 2 possible and we will do the same for other possible duplicates in the table `species_info`.

Drop duplicates with the same `scientific_name` column.

```python
# Drop duplicates
spe_df.drop_duplicates(inplace=True, subset='scientific_name', keep='first')
spe_df.reset_index(drop=True, inplace=True)
```

Create categorial values for the column `conservation_status`. We will use this to calculate the worst case scenario since we can appreciate multiple values from this column.

```python
# Create categorical values for the column conservation_status yo be used later for sorting and data manipulation
spe_df.conservation_status = pd.Categorical(spe_df.conservation_status, 
                                               categories=['No Information', 'Species of Concern', 
                                               'In Recovery', 'Threatened', 'Endangered'], 
                                               ordered=True)
```

Create the worst cases for the column `scientific_name` by selecting the one wit the highest value. We already have created a categorical for the column.

```python
worst_cases = s_duplicates.groupby('scientific_name').conservation_status.max()
worst_cases = worst_cases.sort_values(axis=0, ascending=True)
display(worst_cases)
```

After printing the worst cases we can see the ones with the most endangered `conservation_status` are selected.

Update the species_info clean DataFrame (We eliminated the duplicates) with the worst cases data.

```python
# Update the status for each duplicate that we kept
for species, worst_risk in worst_cases.iteritems():
    spe_df.conservation_status[spe_df.scientific_name == species] = worst_risk
spe_df
```

## Are certain kinds of species more likely to be endangered?

#### Statistical Significance

We will run some chi-squared tests to see if different kinds of species have statistically significant differences in conservation status rates. In order to run a chi squared test, a `contingency` table will need to be created.  

We will use pd.crosstab to create this table using the dataframe created in steps above called `category_all`.

The results from the chi-squared test returns multiple values. The standard threshold to test statistical significance is 0.05. For the value retrieved from this test, the `pval` value of 0.69 is much larger than 0.05. In the case **"if different kinds of species have statistically significant differences in conservation status rates"** there doesn't seem to be any significant relationship 

Are the differences between species and their conservation status significant?
Which species were spotted the most at each park?


```python
# We will run a chi2 contingency
# Load libary 
from scipy.stats import chi2_contingency

# pre-set threshold
significance_threshold = 0.05
# run a chi-square test
contingency = pd.crosstab(spe_df.protected, spe_df.category)
print(contingency)
chi2, pval, dof, expected = chi2_contingency(contingency, correction=False)
# determine significance here:
result = result = ('significant' if pval < significance_threshold else 'not significant')
print('There is ' + result + ' result with pval being '+ str(pval))
```

From the  previous results we can conclude that some categories os species are more likely to be protected than others since the `pval` is < the threshold of `.05`.

## Are the differences between species's categories and their conservation status significant?

First lets create a dataframe with just the 'Protected' species.

```python
clean_spe_df = spe_df[spe_df.protected == 'Protected']
clean_spe_df
```

Create a new contingency table `contingency2` for just the protcted species to calculate the Chi Square.

```python
# Create a new contingency table for just the protected species
contingency2 = pd.crosstab(clean_spe_df.conservation_status, clean_spe_df.category)
print(contingency2)

# run a chi-square test
chi2, pval, dof, expected = chi2_contingency(contingency2, correction=False)
# determine significance here:
result = result = ('significant' if pval < significance_threshold else 'not significant')
print('There is ' + result + ' result with pval being '+ str(pval))
```

The differences between species's categories,  conservation status is significant. `pval` is less then the threshold of 0.05. This means the species do not have similar conservation statuses.
